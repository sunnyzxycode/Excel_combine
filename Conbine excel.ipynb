```python
import os
import pandas as pd

# 文件夹路径
folder_path = r"C:\Users\zxy\Desktop\2022\2022.1"

# 想要提取的列名
columns_to_extract = ['监测点', 'AQI']

# 初始化一个空的DataFrame来存储结果
combined_data = pd.DataFrame(columns=columns_to_extract)

# 遍历文件夹中的每个文件
for file in os.listdir(folder_path):
    if file.endswith('.xlsx'):
        file_path = os.path.join(folder_path, file)
        
        # 读取Excel文件
        df = pd.read_excel(file_path, usecols=columns_to_extract)
        
        # 将数据追加到结果DataFrame中
        combined_data = pd.concat([combined_data, df], ignore_index=True)

# 将合并后的数据保存到新的Excel文件
output_file = 'combined_data.xlsx'
combined_data.to_excel(output_file, index=False)

print("文件合并完成，保存在：", output_file)
```

```python
import os
import pandas as pd

# 文件夹路径
folder_path = r"C:\Users\zxy\Desktop\demo2022"

# 想要提取的列名
columns_to_extract = ['监测点', 'O3']

# 初始化一个空的DataFrame来存储结果
combined_data = pd.DataFrame()

# 遍历文件夹中的每个文件
for file in os.listdir(folder_path):
    if file.endswith('.xlsx'):
        file_path = os.path.join(folder_path, file)

        # 尝试读取Excel文件的前几行来检查列名
        try:
            df_preview = pd.read_excel(file_path, nrows=5)
            if all(col in df_preview.columns for col in columns_to_extract):
                # 如果所需的列都存在，则读取这些列
                df = pd.read_excel(file_path, usecols=columns_to_extract)

                # 获取不包含扩展名的文件名
                file_name = os.path.splitext(file)[0]

                # 为每个文件的数据设置新的列名
                new_column_names = [f'{file_name}_{name}' for name in columns_to_extract]
                df.columns = new_column_names

                # 将数据合并到结果DataFrame中，按列合并
                combined_data = pd.concat([combined_data, df], axis=1)
            else:
                # 如果所需列不存在，打印信息并跳过该文件
                print(f"文件 '{file}' 中缺少所需列，已跳过。")
        except Exception as e:
            print(f"处理文件 '{file}' 时发生错误：{e}")

# 将合并后的数据保存到新的Excel文件
output_file = 'O3.xlsx'

combined_data.to_excel(output_file, index=False)

print("文件合并完成，保存在：", output_file)
```
